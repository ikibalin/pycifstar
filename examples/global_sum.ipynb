# Library PyStar
## Basic operations 

The reading of cif blocks: loop, data, global from file to objects.

```python
import pystar
f_1 = "global_1.rcif"
f_2 = "global_2.rcif"
f_3 = "global_3.rcif"
global_1 = pystar.read_file(f_1)
global_2 = pystar.read_file(f_2)
global_3 = pystar.read_file(f_3)
```

```python
global_123 = global_1+global_2+global_3
```

```python
print(global_123)
```

```python
global_123.to_file("global_sum.rcif")
```
